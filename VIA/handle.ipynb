```python
from pixetto import Pixetto
import time
import requests
```

```python
pix = Pixetto()
pix.open("COM3")
```

```python
count = 0
clock = 0
x, y = [0, 0], [0, 0]
while True:
    time.sleep(0.1)
    try:
        if pix.is_detected() == True:
            _, _, plates = pix.get_data_list()
            
            if (abs(plates[0]['x'] - x[0]) < 30 or abs(plates[0]['x'] - x[1]) < 30) and (abs(plates[1]['x'] - x[1]) < 30 or abs(plates[1]['x'] - x[0]) < 30) and (abs(plates[0]['y'] - y[0]) < 30 or abs(plates[0]['y'] - y[1]) < 30) and (abs(plates[1]['y'] - y[1]) < 30 or abs(plates[1]['y'] - y[0]) < 30):
                count += 1
                print(count)
            else:
                count = 0
            x[0] = plates[0]['x']
            x[1] = plates[1]['x']
            y[0] = plates[0]['y']
            y[1] = plates[1]['y']
            #print(plates[0]['x'], plates[1]['x'], plates[0]['y'], plates[1]['y'])
            #print(x[0], x[1], y[0], y[1])
            if count > 50:
                print('Customers do not want to eat anymore !')
                if clock == 20:
                    print('request eat 0')
                    requests.get('http://140.113.68.171:35000/toolong?eat=0')
                    clock = 0
                    time.sleep(3)
            else:
                if clock == 20:
                    print('request eat 1')
                    requests.get('http://140.113.68.171:35000/toolong?eat=1')
                    clock = 0
                    time.sleep(3)
            clock += 1
            #print(clock)
    except KeyboardInterrupt:
        raise
    except:
        pass
pix.close()
```
